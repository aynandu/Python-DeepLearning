```python
import numpy as np  # For Algebric Expressions
import matplotlib.pyplot as plt  # for Graphical Representation
```

```python
time = np.linspace(0,40,1000) # start , finish , a points
```

Distance travelled by robber ; dst_rob=2.5time

```python
dst_rob =  2.5*time
```

Distance Travelled by Sheriff : dis_sherf = 3(t-5)

```python
dis_sherf=3 *(time-5)
```

```python
fig , ax = plt.subplots()
plt.title("A Bank Robber Caught")
plt.xlabel("time (in minutes)")
plt.ylabel("distance (in km)")
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(time,dst_rob,'green')
ax.plot(time,dis_sherf,"red")
plt.axvline(x=30,color="purple",linestyle='--')
__=plt.axhline(y=75,color="purple",linestyle='--')
```
